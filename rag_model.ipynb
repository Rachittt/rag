```python
from llama_index.core import VectorStoreIndex,SimpleDirectoryReader,ServiceContext
from llama_index.core.prompts.prompts import SimpleInputPrompt
from llama_index.llms.huggingface import HuggingFaceLLM
```

Read all the document in data directory and save it in a variable

```python
documents=SimpleDirectoryReader("./data").load_data()
len(documents)
```

```python
system_prompt="""
You are a Q&A assistant. Your goal is to answer questions as
accurately as possible based on the instructions and context provided.
""" 
## Default format supportable by LLama2
query_wrapper_prompt=SimpleInputPrompt("<|USER|>{query_str}<|ASSISTANT|>")
```

Create a llm model,
Before creating the model you have to take access from meta to use the model

```python
import torch
llm = HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="meta-llama/Llama-2-7b-chat-hf",
    model_name="meta-llama/Llama-2-7b-chat-hf",
    device_map="auto",
    model_kwargs={"torch_dtype": torch.float16 , "load_in_8bit":True}
  )
```

```python
%pip install llama-index-embeddings-langchain
```

```python
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from llama_index.core import ServiceContext
from llama_index.embeddings.langchain import LangchainEmbedding
embed_model=LangchainEmbedding(HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2"))
```

```python
service_context=ServiceContext.from_defaults( chunk_size=1024, llm=llm, embed_model=embed_model )
```

```python
service_context
```

Creating Vectorized version of data

```python
index=VectorStoreIndex.from_documents(documents,service_context=service_context)
```

```python
index
```

Creating a query engine from the vectorized data

```python
query_engine=index.as_query_engine()
```

```python
response1=query_engine.query("Identify the industry in which Canoo operates, along with its size, growth rate, trends, and key players.")
```

```python
print(response1)
```

```python
response2=query_engine.query("Analyze Canoo's main competitors, including their market share, products or services offered, pricing strategies, and marketing efforts.")
print(response2)
```

```python
response3=query_engine.query("Identify key trends in the market, including changes in consumer behavior, technological advancements, and shifts in the competitive landscape.")
print(response3)
```

```python
response4=query_engine.query("Gather information on Canoo's financial performance, including its revenue, profit margins, return on investment, and expense structure.")
print(response4)
```

Insert all the questions user want to ask in the below list

```python
query_list = ["Identify the industry in which Canoo operates, along with its size, growth rate, trends, and key players.",
              "Analyze Canoo's main competitors, including their market share, products or services offered, pricing strategies, and marketing efforts.",
              "Identify key trends in the market, including changes in consumer behavior, technological advancements, and shifts in the competitive landscape.",
              "Gather information on Canoo's financial performance, including its revenue, profit margins, return on investment, and expense structure."]
```

```python
responses={}
for question in query_list:
  response=query_engine.query(question)
  responses[question]=response
```

```python
len(responses)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(list(responses.items()), index=None, columns=['question', 'answer'])
```
